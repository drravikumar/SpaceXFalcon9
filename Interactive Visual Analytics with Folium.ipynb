Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
import folium
from folium.plugins import MarkerCluster
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python
# Assuming spacex_df contains 'Latitude' and 'Longitude' for launch sites
# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
# Create a marker for each launch record
mean_lat = spacex_df['Lat'].mean()
mean_lon = spacex_df['Long'].mean()

site_map = folium.Map(location=[mean_lat, mean_lon], zoom_start=5)
marker_cluster = MarkerCluster().add_to(site_map)

for index, row in spacex_df.iterrows():
    color = 'green' if row['class'] == 1 else 'red'
    folium.Marker(
        location=[row['Lat'], row['Long']],
        popup=f"Launch Site: {row['Launch Site']}<br>Success: {'Yes' if row['class'] == 1 else 'No'}",
        icon=folium.Icon(color=color)
    ).add_to(marker_cluster)

site_map

```

```python
import pandas as pd

# Example DataFrame creation (This should be your actual DataFrame loading step)
# spacex_df = pd.read_csv('path_to_your_spacex_data.csv')

# Define the function to determine the marker color
def determine_color(class_value):
    return 'green' if class_value == 1 else 'red'

# Apply the function to create the marker_color column
spacex_df['marker_color'] = spacex_df['class'].apply(determine_color)

# Display the first few rows to verify the new column
print(spacex_df[['class', 'marker_color']].head())
```

```python
# Add marker_cluster to current site_map
site_map.add_child(marker_cluster)

# Iterate through each row in spacex_df data frame
for index, record in spacex_df.iterrows():
    # Create a Marker object with its coordinate and customize the Marker's icon
    # The color of the icon is determined by the 'marker_color' column
    marker = folium.Marker(
        location=[record['Lat'], record['Long']],
        popup=f"Launch Site: {record['Launch Site']}<br>Success: {'Yes' if record['class'] == 1 else 'No'}",
        icon=folium.Icon(color=record['marker_color'], icon_color='white')  # Assuming 'white' is the desired icon color within the button
    )
    # Add the Marker to the MarkerCluster
    marker_cluster.add_child(marker)

# Show the map with all markers
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
